```python
feature = ['price','bedrooms','bathrooms','sqft_living','view','grade','sqft_above','yr_built','yr_renovated','sqft_living15']

house_df = house[feature]
```

```python
# Relationships between variables

sns.pairplot(house_df)
```

```python
# Set Seaborn style
sns.set_style("darkgrid")
 
# Identify numerical columns
numerical_columns = house_df.select_dtypes(include=["int64", "float64"]).columns
 
# Plot distribution of each numerical feature
plt.figure(figsize=(14, len(numerical_columns) * 3))
for idx, feature in enumerate(numerical_columns, 1):
    plt.subplot(len(numerical_columns), 2, idx)
    sns.histplot(house_df[feature], kde=True)
    plt.title(f"{feature} | Skewness: {round(house_df[feature].skew(), 2)}")

plt.tight_layout()
plt.show()
```

```python
# Define x and y variables

new_feature = ['bedrooms','bathrooms','sqft_living','view','grade','sqft_above','yr_built','yr_renovated','sqft_living15']

x = house_df[new_feature]
y = house_df['price']
```

```python
x.head()
```

```python
y.head()
```

```python
# Setting up the model

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

train_x,val_x,train_y,val_y = train_test_split(x,y)
linear_model = LinearRegression()
linear_model.fit(train_x,train_y)
```

```python
train_x
```

```python
train_y
```

```python
val_x
```

```python
val_y
```

```python
# Model Evaluation

print(linear_model.intercept_)
```

```python
coeff_df = pd.DataFrame(linear_model.coef_, x.columns, columns=['Coefficient'])
coeff_df
```

```python
# Make predictions from model

predictions = linear_model.predict(val_x)
plt.scatter(val_y, predictions)
```

```python
sns.displot((val_y-predictions), bins=20)
```

```python
# Evaluations

from sklearn import metrics

print('MAE:', metrics.mean_absolute_error(val_y, predictions))
print('MSE:', metrics.mean_squared_error(val_y, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(val_y, predictions)))
```
